```python
i = 0
Long_column_name = []
for column in tqdm(data.columns):
    explanation_of_columns = data2['QuestionText']
    print('\n\n',column,':--',explanation_of_columns[i])
    print("Value_counts for the column of :",column)
    if (len(data[column].value_counts()) >10):
        print("Greater than 10")
        Long_column_name.append(column)
    else:
        print(data[column].value_counts())
    i=i+1
```

```python
data.drop(columns={'Gender','Trans','Sexuality','Ethnicity','Dependents','SurveyLength','SurveyEase','SONewContent','WelcomeChange','SOComm','SOPartFreq','SOAccount','SOHowMuchTime','SOTimeSaved','SOFindAnswer','SOVisitTo','SOVisitFreq','SOVisit1st','ScreenName','Extraversion','OffOn','BlockchainIs','BlockchainOrg','Containers','WorkRemote','CompFreq','CompTotal','OpenSource','Student','EduOther','Age1stCode','MgrIdiot','MgrMoney','MgrWant','LastHireDate','LastInt','FizzBuzz','JobFactors','ResumeUpdate','CurrencySymbol','CurrencyDesc','CompTotal'},inplace = True)
```

```python
MainBranch = []
for i in data['MainBranch']:
    if i == 'I am a developer by profession':
        MainBranch.append('Developer')
    elif i == 'I am a student who is learning to code':
        MainBranch.append('Student')
    elif i == 'I am not primarily a developer, but I write code sometimes as part of my work':
        MainBranch.append('Semi_Developer')
    elif i == 'I code primarily as a hobby':
        MainBranch.append('Hobby') 
    elif i == 'I used to be a developer by profession, but no longer am':
        MainBranch.append('Ex_Developer')
    else:
        MainBranch.append('')
data['MainBranch'] = MainBranch
```

```python
OpenSourcer = []
for i in data['OpenSourcer']:
    if (i == 'Never' ) or (i=='Less than once per year') or (i==''):
        OpenSourcer.append('No')
    else:
        OpenSourcer.append('Yes')
data['OpenSourcer'] = OpenSourcer
```
